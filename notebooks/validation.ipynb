```python
import sys
sys.path.append("../src")
```

```python
import pandas as pd

import readers as readers

DATA_DIR = "../assets/data"
SCHEMAS = {}
```

```python
dfs = readers.load_csv(DATA_DIR, enforce_schema=False)
dfs.keys()
```

# Products

```python
products = dfs["products"]
products.info()
```

```python
products.set_index("product_id", inplace=True)
```

```python
SCHEMAS["products"] = products.dtypes.apply(str).to_dict()
SCHEMAS["products"]
```

# Customers

```python
customers = dfs["customers"]
customers.info()
```

```python
customers.set_index("customer_id", inplace=True)
customers = customers.astype({"signup_date": "datetime64[s]"})
```

```python
SCHEMAS["customers"] = customers.dtypes.apply(str).to_dict()
SCHEMAS["customers"]
```

# Promotions

```python
promotions = dfs["promotions"]
promotions.info()
```

```python
promotions.set_index("promo_id", inplace=True)
promotions = promotions.astype({
    "start_date": "datetime64[s]",
    "end_date": "datetime64[s]"
})
```

```python
SCHEMAS["promotions"] = promotions.dtypes.apply(str).to_dict()
SCHEMAS["promotions"]
```

# Geography

```python
geography = dfs["geography"]
geography.info()
```

```python
geography.set_index("zip", inplace=True)
```

```python
SCHEMAS["geography"] = geography.dtypes.apply(str).to_dict() 
SCHEMAS["geography"]
```

# Orders

```python
orders = dfs["orders"]
orders.info()
```

```python
orders.set_index("order_id", inplace=True)
orders = orders.astype({"order_date": "datetime64[s]"})
```

```python
SCHEMAS["orders"] = orders.dtypes.apply(str).to_dict()  
SCHEMAS["orders"]
```

# Order Items

```python
SCHEMAS["order_items"] = order_items.dtypes.apply(str).to_dict()
SCHEMAS["order_items"]
```

# Payments

```python
payments = dfs["payments"]
payments.info()
```

```python
payments.set_index("order_id", inplace=True)
```

```python
SCHEMAS["payments"] = payments.dtypes.apply(str).to_dict()
SCHEMAS["payments"]
```

# Shipments

```python
shipments = dfs["shipments"]
shipments.info()
```

```python
shipments.set_index("order_id", inplace=True)
shipments = shipments.astype({
    'ship_date': "datetime64[s]",
    'delivery_date': "datetime64[s]"
})
```

```python
SCHEMAS["shipments"] = shipments.dtypes.apply(str).to_dict() 
SCHEMAS["shipments"]
```

# Returns

```python
returns = dfs["returns"]
returns.info()
```

```python
returns.set_index("return_id", inplace=True)
returns = returns.astype({"return_date": "datetime64[s]"})
```

```python
SCHEMAS["returns"] = returns.dtypes.apply(str).to_dict()  
SCHEMAS["returns"]
```

# Reviews

```python
reviews = dfs["reviews"]
reviews.info()
```

```python
reviews.set_index("review_id", inplace=True)
reviews = reviews.astype({"review_date": "datetime64[s]"})
```

```python
SCHEMAS["reviews"] = reviews.dtypes.apply(str).to_dict()
SCHEMAS["reviews"]
```

# Sales

```python
sales = dfs["sales"]
sales.info()
```

```python
sales = sales.astype({"Date": "datetime64[s]"})
```

```python
SCHEMAS["sales"] = sales.dtypes.apply(str).to_dict() 
SCHEMAS["sales"]
```

# Inventory

```python
inventory = inventory.astype({"snapshot_date": "datetime64[s]"})
```

# Web Traffic
